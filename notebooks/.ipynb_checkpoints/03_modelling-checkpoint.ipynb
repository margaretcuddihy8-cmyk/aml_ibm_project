# Modelling

This notebook builds and evaluates classification models to detect potential laundering activity using a feature-engineered transaction dataset. It applies a reproducible 60/20/20 split for training, validation, and testing, ensuring stratified class balance across all sets.

Two supervised learning algorithms are implemented:
- Random Forest for baseline performance and feature importance
- XGBoost for optimized gradient boosting and fine-tuned control
  
Model performance is assessed using appropriate metrics for imbalanced classification, including:
- Precision
- Recall
- F1-score
- ROC-AUC


## 1. Imports and Setup
Import core libraries for data manipulation and visualisation. Seaborn is used for styling plots.

```python
import sys, os
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), "..")))

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import StratifiedKFold, cross_val_score, cross_validate, RandomizedSearchCV
from sklearn.metrics import confusion_matrix, classification_report, precision_recall_curve
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.feature_selection import VarianceThreshold
from sklearn.pipeline import Pipeline


from xgboost import XGBClassifier

import shap

from src.data_loader import load_transaction_data
from src.feature_engineering import engineer_features
from src.model_utils import stratified_split, check_class_balance, shap_case_summary
from src.report_utils import get_modeling_summary
from src.report_utils import summarise_features
```

## 2. Load Feature-Engineered Data

```python
df = load_transaction_data("../data/raw/HI-Small_Trans.csv.gz", verbose=False)
df_trans = engineer_features(df)
```

## 3. Define target variable

```python
# Define target
y = df_trans['is_laundering']

# Define features 
X = df_trans.drop(columns=['is_laundering'])

# Drop non-predictive columns
X = df_trans.drop(columns=['is_laundering', 'timestamp', 'account', 'account.1'])
```

## 4. Split Data into Train and Test Sets
Stratified split using 70/30 split.

```python
# 70% train, 30% test stratified split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, stratify=y, random_state=42)
```

```python
# Check class distribution after splitting
check_class_balance(y_train, "Training")
check_class_balance(y_test, "Test")
```

```python
summary_df = summarise_features(df_trans)
print(summary_df.to_markdown(index=False))
```

## 4. XGBoost Model

```python
# Select only numeric columns
X_train = X_train.select_dtypes(include=["number", "bool"])
X_test = X_test.select_dtypes(include=["number", "bool"])

# Checked variances to include initial pruning of features with low variance. 
selector = VarianceThreshold(threshold=0.0)
selector.fit(X_train)
variances = pd.Series(selector.variances_, index=X_train.columns)
# print(variances.sort_values())
```

```python
# Calculate class imbalance ratio
scale = (y_train == 0).sum() / (y_train == 1).sum()

# Define pipeline
pipeline = Pipeline([
    ('var_thresh', VarianceThreshold(threshold=0.02)),
    ('model', XGBClassifier(
        scale_pos_weight=scale,
        base_score=0.5,
        eval_metric='logloss',
        random_state=42,
        n_jobs=-1
    ))
])

# Quick param grid
param_dist = {
    'model__n_estimators': [25, 50, 100],
    'model__max_depth': [3, 5, 7],
    'model__learning_rate': [0.01, 0.1, 0.3]
}

# Randomized search (quick: 5 iterations)
search = RandomizedSearchCV(
    pipeline,
    param_distributions=param_dist,
    n_iter=5,
    scoring='f1',
    cv=3,
    random_state=42,
    n_jobs=-1,
    verbose=1
)

# Fit model
search.fit(X_train, y_train)

# Best model
best_model = search.best_estimator_
print("Best parameters:", search.best_params_)
```

```python
# Evaluate the best model from tuning

cv = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)

scores = cross_val_score(search.best_estimator_, X_train, y_train, cv=cv, scoring='f1', n_jobs=-1)
print("Tuned XGBoost F1 scores:", scores)
print("Mean F1:", scores.mean())
```

```python
# Align test set to training set columns
X_test = X_test[X_train.columns]

# Now predict
y_pred = best_model.predict(X_test)

# Confusion matrix
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred, digits=4))
```

```python
cm = confusion_matrix(y_test, y_pred)
labels = ['True Negatives', 'False Positives', 'False Negatives', 'True Positives']

plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Pred 0', 'Pred 1'], yticklabels=['Actual 0', 'Actual 1'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

Accuracy is high because of the high number of negatives. The recall score for flagged transactions shows the model manages to identify 94% of true flagged cases. Precision is extremely low as most of the flagged cases have been incorrectly identified for flagging. It depends on the cost of failing to identify suspicious banking activity versus the cost of needlessly investigating lawful transactions.

```python
y_proba = best_model.predict_proba(X_test)[:, 1]

precision, recall, thresholds = precision_recall_curve(y_test, y_proba)

plt.figure(figsize=(8, 6))
plt.plot(thresholds, precision[:-1], label='Precision')
plt.plot(thresholds, recall[:-1], label='Recall')
plt.xlabel('Threshold')
plt.ylabel('Score')
plt.title('Precision-Recall vs Threshold')
plt.legend()
plt.grid(True)
plt.show()
```

## False Negatives Analysis
This represents a higher risk as missing transactions that should be flagged means that AML can continue undedected.

```python
# Examine false negatives

# Reset index to align with y_pred
y_test_reset = y_test.reset_index(drop=True)

# Create results DataFrame
results = pd.DataFrame({
    'true_label': y_test_reset,
    'predicted_label': y_pred
})

# Identify false negatives
false_negatives = results[(results['predicted_label'] == 0) & (results['true_label'] == 1)]
print(f"False negatives: {len(false_negatives)}")

# Use .iloc to extract rows by position
false_negative_cases = X_test.iloc[false_negatives.index].copy()
false_negative_cases.reset_index(drop=True, inplace=True)

# Now this assignment will align correctly
false_negative_cases['true_label'] = y_test_reset.iloc[false_negatives.index].values
false_negative_cases['predicted_label'] = y_pred[false_negatives.index]
```

```python
false_negative_cases.head()
```

```python
fn_numeric = false_negative_cases.select_dtypes(include=[np.number])
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(fn_numeric)

false_negative_cases['cluster'] = clusters
false_negative_cases.groupby('cluster').mean(numeric_only=True)
```

```python
pca = PCA(n_components=2)
components = pca.fit_transform(fn_numeric)

plt.figure(figsize=(8, 6))
sns.scatterplot(x=components[:, 0], y=components[:, 1], hue=clusters, palette='Set2')
plt.title("False Negative Clusters (PCA)")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.legend(title='Cluster')
plt.show()
```

```python
def compare_density_fn_tn(feature_name, X_test, results, figsize=(10, 6)):
    false_negatives = results[(results['predicted_label'] == 0) & (results['true_label'] == 1)].index
    true_negatives = results[(results['predicted_label'] == 0) & (results['true_label'] == 0)].index

    fn_values = X_test.iloc[false_negatives][feature_name].dropna()
    tn_values = X_test.iloc[true_negatives][feature_name].dropna()

    plt.figure(figsize=figsize)
    sns.kdeplot(fn_values, label='False Negatives', color='red', fill=True)
    sns.kdeplot(tn_values, label='True Negatives', color='gray', fill=True)
    plt.xlabel(feature_name)
    plt.ylabel('Density')
    plt.title(f'{feature_name}: False Negatives vs True Negatives')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
for feature in X_test.select_dtypes(include=np.number).columns:
    compare_density_fn_tn(feature, X_test, results)
```

## False Positives Analysis
While not as high risk as false negatives, the majority of flagged predictions are incorrect and it's still a drain on resources to investigate legitimate transactions uneccessarily.

```python
# Examine false negatives

# Reset index to align with y_pred
y_test_reset = y_test.reset_index(drop=True)

# Create results DataFrame
results = pd.DataFrame({
    'true_label': y_test_reset,
    'predicted_label': y_pred
})

# Identify false positives
false_positives = results[(results['predicted_label'] == 1) & (results['true_label'] == 0)]
print(f"False positives: {len(false_positives)}")

# Use .iloc to extract rows by position
false_positive_cases = X_test.iloc[false_positives.index].copy()
false_positive_cases.reset_index(drop=True, inplace=True)

# Now this assignment will align correctly
false_positive_cases['true_label'] = y_test_reset.iloc[false_positives.index].values
false_positive_cases['predicted_label'] = y_pred[false_positives.index]
```

```python
false_positive_cases.head()
```

```python
fp_numeric = false_positive_cases.select_dtypes(include=[np.number])
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(fp_numeric)

false_positive_cases['cluster'] = clusters
```

```python
false_positive_cases.groupby('cluster').mean(numeric_only=True)
```

```python
pca = PCA(n_components=2)
components = pca.fit_transform(fp_numeric)

plt.figure(figsize=(8, 6))
sns.scatterplot(x=components[:, 0], y=components[:, 1], hue=clusters, palette='Set2')
plt.title("False Positive Clusters (PCA)")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.legend(title='Cluster')
plt.show()
```

```python
def compare_density_fp_tp(feature_name, X_test, results, figsize=(10, 6)):
    """
    Compare the distribution of a feature between false positives and true positives.

    Parameters:
    - feature_name: str, name of the feature to compare
    - X_test: pd.DataFrame, encoded validation features
    - results: pd.DataFrame with 'true_label' and 'predicted_label'
    - figsize: tuple, size of the plot
    """
    # Identify false and true positives
    false_positives = results[(results['predicted_label'] == 1) & (results['true_label'] == 0)].index
    true_positives = results[(results['predicted_label'] == 1) & (results['true_label'] == 1)].index

    # Extract feature values
    fp_values = X_test.iloc[false_positives][feature_name]
    tp_values = X_test.iloc[true_positives][feature_name]

    # Plot density comparison
    plt.figure(figsize=figsize)
    sns.kdeplot(fp_values, label='False Positives', color='blue', fill=True)
    sns.kdeplot(tp_values, label='True Positives', color='green', fill=True)
    plt.xlabel(feature_name)
    plt.ylabel('Density')
    plt.title(f'{feature_name}: False Positives vs True Positives')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
for feature in X_test.select_dtypes(include=np.number).columns:
    compare_density_fp_tp(feature, X_test, results)
```

## 7. SHAP Analysis

```python

def plot_shap_summary_for_cases(
    case_mask,
    X_test,
    y_test,
    X_train_enc,
    pipeline,
    case_label="false_negatives",
    sample_size=100
):
    """
    Generates and saves a SHAP summary plot for selected cases (e.g., false negatives).

    Parameters:
    - case_mask (pd.Series or np.array): Boolean mask for selecting cases.
    - X_test (pd.DataFrame): Encoded validation features.
    - y_test (pd.Series or np.array): Validation labels.
    - X_train_enc (pd.DataFrame): Encoded training features for column alignment.
    - pipeline (sklearn.Pipeline): Fitted pipeline with 'var_thresh' and 'model' steps.
    - case_label (str): Label for saving and logging (e.g., "false_negatives").
    - sample_size (int): Max number of samples to use for SHAP.

    Returns:
    - None. Displays and saves SHAP summary plot.
    """

    # Extract model and selector
    pipeline = Pipeline([
    ('var_thresh', VarianceThreshold(threshold=0.0)),
    ('model', best_model)
    ])

    pipeline.fit(X_train, y_train)


    # Select cases
    case_df = X_test[case_mask]
    n_cases = len(case_df)
    if n_cases == 0:
        print(f"[SHAP] No cases found for label: {case_label}")
        return

    sample_n = min(sample_size, n_cases)
    case_sample = case_df.sample(sample_n, random_state=42)

    # Align columns
    case_clean = case_sample[X_train_enc.columns]
    case_reduced = selector.transform(case_clean)
    selected_features = X_train_enc.columns[selector.get_support()]
    case_df_reduced = pd.DataFrame(case_reduced, columns=selected_features)

    # SHAP explanation
    explainer = shap.TreeExplainer(best_model)
    shap_values = explainer.shap_values(case_df_reduced)

    # Plot
    os.makedirs("outputs/plots", exist_ok=True)
    print(f"[SHAP] Summary for: {case_label} ({sample_n} of {n_cases} cases)")
    plt.figure()
    shap.summary_plot(shap_values, case_df_reduced, show=True)
    plt.tight_layout()
    plot_path = f"outputs/plots/shap_summary_{case_label}.png"
    plt.savefig(plot_path)
    plt.close()
    print(f"[SHAP] Plot saved to: {plot_path}")

```

```python
false_negatives = (y_test.values.flatten() == 1) & (y_pred == 0)
plot_shap_summary_for_cases(false_negatives, X_test, y_test, X_train, pipeline, case_label="False Negatives")

true_negatives = (y_test.values.flatten() == 0) & (y_pred == 0)
shap_case_summary(true_negatives, X_test, y_test, X_train, pipeline, case_label="True Negatives")

false_positives = (y_test.values.flatten() == 0) & (y_pred == 1)
shap_case_summary(false_positives, X_test, y_test, X_train, pipeline, case_label="False Positives")

true_positives = (y_test.values.flatten() == 1) & (y_pred == 1)
shap_case_summary(true_positives, X_test, y_test, X_train, pipeline, case_label="True Positives")
```

SHAP analysis reveals that sender_freq is the most influential feature in predicting laundering risk, followed by transaction format and behavioral patterns such as off-hour activity and weekday timing. Engineered interaction terms further enhance model sensitivity, confirming the value of behavioral feature engineering in forensic detection.
